# **심화 기능(Advanced-Features)**

## 표준 라이브러리(Standard Libraries)
* **표준 라이브러리(Standard Libraries)** 파이썬에 기본 내장되어 별도의 설치 없이 바로 사용할 수 있는 강력한 도구 모음

* **OS, 파일 시스템(`os`, `glob`, `shutil`)** 운영체제와 상호작용하며 파일과 폴더 관리
  * **`os` 모듈** 시스템 경로 및 프로세스 관리
    * **`os.path.join(path, *paths)`** 운영체제별 구분자(`/` 또는 `\`)에 맞춰 경로를 안전하게 결합
    * **`os.path.exists(path)`** 경로 존재 여부 확인(`True`/`False`)
    * **`os.path.isfile(path)` / `os.path.isdir(path)`** 파일인지 디렉토리인지 확인
    * **`os.makedirs(path, exist_ok=True)`** 폴더 생성(중간 단계 폴더 포함, 이미 있으면 무시)
    * **`os.remove(path)`** 파일 삭제(폴더 삭제 불가)
  * **`glob` 모듈** 파일 패턴 매칭
    * **`glob.glob(pathname)`** `*`, `?` 와일드카드를 사용해 일치하는 파일 목록을 리스트로 반환.
  * **`shutil` 모듈** 고급 파일 조작
    * **`shutil.copy(src, dst)`** 파일 복사(권한 복사 O, 메타데이터 X)
    * **`shutil.move(src, dst)`** 파일/폴더 이동 또는 이름 변경(이동 후 원본 삭제)
    * **`shutil.rmtree(path)`** 폴더 및 하위 모든 파일/폴더 강제 일괄 삭제
* **수학 및 랜덤(`math`, `random`)**
  * **`math` 모듈** 고수준 수학 연산
    * 상수 = `math.pi($\pi \approx 3.14$)`, `math.e($e \approx 2.71$)`
    * 함수 = `math.log(x, [base])`(로그), `math.sin/cos/tan(x)`(삼각함수, 라디안 기준)
    * 각도(`degree`) 사용 시 `math.radians(90)`으로 변환 필요
  * **`random` 모듈** 무작위 값 생성
    * **`random.random()`** 0.0 ~ 1.0 미만 실수
    * **`random.randint(a, b)`** $a \le x \le b$ 범위의 정수
    * **`random.uniform(a, b)`** $a \le x \le b$ 범위의 실수
    * **`random.shuffle(list)`** 리스트 순서 무작위 섞기 (원본 변경).
    * **`random.choice(list)`** 요소 중 하나 무작위 선택
    * **`random.sample(list, k)`** 중복 없이 $k$개 추출
* **시간 관리(`time`, `datetime`)**
  * **`time.time()`** Unix Timestamp(1970년 이후 경과 초) 반환
  * **`time.sleep(secs)`** 지정한 초만큼 실행 일시 정지
  * **`datetime.datetime.now()`** 현재 지역의 년/월/일/시/분/초/마이크로초 정보 객체 반환

  | 모듈 | 용도 | 주요 함수 및 특징 |
  | :--- | :--- | :--- |
  | **os** | 운영체제 및 파일 시스템 제어 | `join()` (경로 결합), `exists()` (존재 확인), `makedirs()` (폴더 생성), `remove()` (파일 삭제) |
  | **math** | 수학적 계산 | `pi`, `e` (상수), `log()`, `sin/cos/tan()`, `radians()` (각도 변환) |
  | **random** | 무작위 값 생성 | `randint()` (정수), `uniform()` (실수), `shuffle()` (섞기), `choice/sample()` (선택/추출) |
  | **glob** | 파일 패턴 매칭 | `glob()`: 와일드카드(`*`, `?`)를 사용해 파일 목록을 리스트로 반환 |
  | **shutil** | 고수준 파일 조작 | `copy()` (복사), `move()` (이동/이름 변경), `rmtree()` (폴더 강제 삭제) |
  | **time & datetime** | 시간 및 날짜 관리 | `time()` (타임스탬프), `sleep()` (지연), `now()` (현재 일시 객체) |

## 효율적인 자원 관리 및 구조
* **제너레이터(Generator)** 모든 값을 한꺼번에 메모리에 올리지 않고 호출될 때마다 값을 하나씩 생성하는 방식

  * Lazy Evaluation(게으른 연산) ➡ 메모리 사용량 최소화


  | 구분 | 일반 함수 (`return`) | 제너레이터 (`yield`) |
    | :--- | :--- | :--- |
    | **반환 방식** | 모든 데이터를 한 번에 반환 | 데이터를 한 줄(요소)씩 생성하여 반환 |
    | **메모리 사용** | 리스트가 커질수록 메모리 점유율 급증 | 데이터 양과 상관없이 낮은 메모리 유지 |
    | **상태 저장** | 실행 후 함수 종료 (상태 소멸) | 마지막 호출 위치를 기억 (상태 유지) |
    | **객체 타입** | 값(Value) 또는 리스트(List) 등 반환 | 반복 가능한 Iterator 객체 반환 |

* **`with`문과 컨텍스트 매니저(Context Manager)**
  * 시작(Setup)과 종료(Teardown)가 쌍으로 이루어지는 작업을 안전하게 처리함

  * 자원(파일, DB 등)의 시작(Setup)과 종료(Teardown) 자동 처리

  * 파일 입출력, DB 연결, 락(Lock) 해제, 시간 측정 등의 용도

  * 실행 도중 에러 발생 시 `close()` 등을 통해 메모리 누수 방지

  *  `@contextmanager` 데코레이터
      * 함수 형식으로 간편하게 생성
      * **`with`문** 자원의 생성과 반납을 자동으로 관리하여 코드를 안전하게 만듦
      * **`yield`** 컨텍스트 매니저(Context Manager) 내부에서 실제 작업이 수행될 위치 지정
      * `yield` 이전은 시작 코드, 이후는 종료 코드 실행

## 디버깅(Debugging) 및 예외 처리
* **디버깅(Debugging)** 프로그램 코드에 있는 오류를 발견하고 수정하는 작업
* 에러 제어 및 예외 처리
  * 에러 강제 발생
    * **`assert` 조건식, 에러 메시지** 조건이 `False`면 `AssertionError` 발생
    * **`raise` 에러 메시지** 의도적 에러 발생 (예: `NotImplementedError`)
    * `try-except` 구조
      * **`try`** 실행할 코드
      * **`except [Exception as e]`** 에러 발생 시 처리(`e`로 에러 정보 출력 가능)
      * **`else`** 에러가 발생하지 않았을 때만 실행
      * **`finally`** 에러 여부와 관계없이 무조건 실행

* **디버깅 도구(`breakpoint()`)** 프로그램을 멈추고 상호작용하는 디버거(pdb) 실행
  * 주요 명령어
    * **`p(`print`)** 변수 값 출력
    * **`n`(`next`)** 현재 줄 실행 후 다음 줄로 이동
    * **`c`(`continue`)** 남은 코드 쭉 실행
    * **`l`(`list`)** 현재 실행 중인 파일의 소스 코드 확인
    * **`q`(`quit`)** 프로그램 종료
