## 나만의 미니 카페 키오스크 만들기
1. 클래스 이름 `CafeKiosk`

2. 변수 조건(Attributes)

* `_menu`: 카페의 메뉴 이름(예: "아이스 아메리카노") - 보호 변수(`_`)

* `__price`: 메뉴의 가격(예: 4500) - 비밀 변수(`__`)

* `is_ice`: 아이스인지 핫인지 결정하는 변수(`True`면 아이스, `False`면 핫)

3. 메서드 조건(Methods)

* `__init__`: 메뉴, 가격, 아이스 여부 저장

* `show_receipt`: 영수증 출력 함수

  * 메뉴 이름 앞에 `[ICE]` 또는 `[HOT]`을 붙여 출력

  * 가격은 클래스 안에서 `self.__price`를 사용해 출력

* `change_price`: 새로운 가격을 인자로 받아 가격 업데이트 함수

4. 조건문
* 가격을 바꿀 때, 새로운 가격이 0원보다 작으면 "가격을 잘못 입력하셨습니다"라고 출력하고 바꾸지 않도록 if문 작성

```python
class CafeKiosk:
  def __init__(self, menu, price, is_ice):
    self._menu = menu
    self.__price = price
    self.is_ice = is_ice
  def show_receipt(self):
    if self.is_ice == True:
      print(f"[ICE]{self._menu} - {self.__price}원")
    else:
      print(f"[HOT]{self._menu} - {self.__price}원")
  def change_price(self, new_price):
    if new_price < 0:
      print("변경 불가")
    else:
      self.__price = new_price
      print(f"가격이 {new_price}원으로 변경되었습니다.")

order1 = CafeKiosk("라떼", 5000, True)
order1.show_receipt()

order1.change_price(5500)
order1.show_receipt()
```

## 전설의 용사 시스템 만들기

1. 부모 클래스 `Character`
* 변수

  * `_name`: 캐릭터 이름(보호 변수)

  * `__hp`: 체력(비밀 변수, 기본값 100)

* 메서드

  * `__init__`: 이름을 받아서 저장

  * `attack`: "공격을 시도합니다!"라고 출력

  * `@property`를 사용해서 비밀 변수인 `__hp`를 읽을 수 있게 만들기

2. 자식 클래스-1 `Warrior`(`Character`를 상속받음)
  * 부모의 `attack`을 무시하고 "칼을 휘둘러 묵직한 일격을 날립니다!"라고 출력하도록 변경
  * drink_potion: "포션을 마십니다!"라고 출력하고, 비밀 변수인 체력을 20만큼 회복시켜야 해.

4. 자식 클래스-2 `Wizard`(`Character`를 상속받음)
  * `attack`을 "지팡이를 휘둘러 강력한 마법을 쏩니다!"라고 출력하도록 변경

5. 실행 및 검증
  * `Warrior` 인스턴스 하나 만들기

  * `attack()`을 호출해서 잘 바뀌었는지 확인

  * `rink_potion()`을 호출한 뒤, 체력이 120이 되었는지 `@property`로 확인

```python
class Character:
  def __init__(self, name, hp=100):
    self._name = name
    self.__hp = hp
  def attack(self):
    print("공격을 시도합니다!")
  @property
  def hp(self):
    return self.__hp
  @hp.setter
  def hp(self, value):
    self.__hp = value

class Warrior(Character):
  def attack(self):
    print("칼을 휘둘러 묵직한 일격을 날립니다!")
  def drink_potion(self):
      print("포션을 마십니다!")
      self.hp = self.hp + 20

class Wizzard(Character):
  def attack(self):
    print("지팡이를 휘둘러 강력한 마법을 쏩니다!")

warrior = Warrior("아이언맨")
warrior.attack()
warrior.drink_potion()
print(f"{warrior._name}의 현재 체력: {warrior.hp}")
```

## 지능형 도서관 대출 시스템

1. 부모 클래스 `Item`(도서관 물품)
* 변수
  * `_title`: 제목(보호 변수)

  * `__is_borrowed`: 대출 여부(비밀 변수, 기본값 `False`)

* 메서드

  * `__init__`: 제목을 받아서 저장

  * `@property`로 `status`: 대출 중이면 "대출 불가", 아니면 "대출 가능"을 반환

  * `borrow()`: 대출 시도, 이미 대출 중이면 "이미 빌려갔어요!" 출력, 아니면 `__is_borrowed`를 `True`로 변경

2. 자식 클래스-1 `Book`(`Item` 상속)
* 변수
  * `author`(작가) 추가

  * `borrow()`를 실행할 때 "책 [제목]을 대출합니다. 즐거운 독서 되세요!"라고 출력하게 업그레이드

3. 자식 클래스-2 `DVD`(`Item` 상속)
* `borrow()`를 실행할 때 "DVD [제목]을 대출합니다. 영화 관람 에티켓을 지켜주세요!"라고 출력

4. 관리 클래스 `Library`
* 변수
  * `__items`: 도서관이 가진 물품 리스트(비밀 변수, 리스트 형태)

* 메서드

  * `__init__`: 빈 리스트로 시작

  * `add_item(item)`: 리스트에 새 물품 추가

  * `show_all()`: 현재 도서관의 모든 물품 이름과 대출 상태(status)를 한눈에 보여주기

```python
class Item:
  def __init__(self, title, is_borrowed = False):
    self._title = title
    self.__is_borrowed = is_borrowed
  @property
  def status(self):
    if self.__is_borrowed == True:
      print("대출 불가")
    else:
      print("대출 가능")
  def borrow(self):
    if self.__is_borrowed == True:
      print("이미 빌려갔어요!")
      return False
    else:
      self.__is_borrowed = True
      return True

class Book(Item):
  def __init__(self, title, author):
    super().__init__(title)
    self.author = author
  def borrow(self):
    if super().borrow():
      print(f"책 [{self._title}]-[{self.author}]을 대출합니다.")

class DVD(Item):
  def borrow(self):
    if super().borrow():
      print(f"DVD [{self._title}]을 대출합니다.")

class Library:
  def __init__(self):
    self._items = []
  def add_item(self, item):
    self._items.append(item)

  def show_all(self):
    for item in self._items:
      print(f"[{item._title}] | [{item.status}]")

lib = Library()
b1 = Book("해리포터", "조앤 롤링")
d1 = DVD("어벤져스")

lib.add_item(b1)
lib.add_item(d1)

lib.show_all() # 목록 확인
b1.borrow()    # 책 대출
b1.borrow()    # 중복 대출 시도
lib.show_all() # 바뀐 상태 확인
```
