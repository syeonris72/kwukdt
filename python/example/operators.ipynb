문제 2) 아래의 문자열 및 리스트의 연산 실행 시 출력되는 결과를 맞혀 보자

```python
print("Hello" + "World") # HelloWorld
```

```python
print("Hello" * 2) # HelloHello
```

```python
print("Hello" + "World" * 2) # HelloWorldWorld
```

```python
print([1, 2] + [1, 1]) # 1, 2, 1, 1
```

```python
print([1, 2] * 2) # 1, 2, 1, 2
```

```python
print([1, 2] + [1, 1] * 2) # 1, 2, 1, 1, 1, 1
```
