# **선택 구조(Selection Structure)**

## 비교 연산자 (Comparison Operators)
* 두 값의 크기를 비교하여 참(`True`) 또는 거짓(`False`)을 결과로 반환하는 연산자
* **주요 연산자**
    * `==` 같다
    * `!=` 다르다
    * `>` / `<` 크다 / 작다
    * `>=` / `<=` 크거나 같다 / 작거나 같다

## random 라이브러리
* `import random`

  `random.함수이름()`
* `randint(시작 값, 끝 값)` 정수 중 하나를 무작위로 반환

```python
import random
num1 = random.randint(1, 10)
answer = int(input('1에서 10의 숫자 중 하나 고르기'))
print(f'결과는 {num1 == answer}, 정답은 {num1}')
```

```python
import random
a = random.randint(10, 50)
b = random.randint(2, 9)
my_quotient = int(input(f'{a} 나누기 {b}의 몫을 예측해보세요! '))
my_remainder = int(input(f'{a} 나누기 {b}의 나머지를 예측해보세요! '))
print(f'몫은 {a // b}로 {a // b == my_quotient}, 나머지는 {a % b}로 {a % b == my_remainder}')
```

## 선택 구조(Selection Structure)
* 조건식의 결과에 따라 실행할 문장을 결정하는 구조
* **if 문** 조건이 참일 때만 실행할 문장 작성
* **if-else 문** 조건이 참일 때와 거짓일 때 실행할 문장을 각각 나누어 작성
* **if-elif-else 문** 여러 개의 조건을 순차적으로 체크해야 할 때 사용
* **참고**

  조건문 끝에 **콜론(:)** 필수

  실행할 문장은 **들여쓰기(Indentation)** 필수

## 논리 연산자(Logical Operators)
* 여러 개의 조건식을 조합하여 하나의 결과를 얻을 때 사용
* **종류**
    * `and` 두 조건이 모두 참일 때만 참
    * `or` 두 조건 중 하나만 참이어도 참
    * `not` 참을 거짓으로, 거짓을 참으로 반전

## 함수(Function)
* 특정 작업을 수행하는 명령어들의 묶음
* **라이브러리 활용** `import` 키워드를 사용하여 외부 도구(라이브러리)를 불러오기 가능
`random.randint(start, end)` 지정한 범위 내의 정수 중 하나를 무작위로 반환

## **실습 예제**




### 온도 변환기 (온도 종류 통합형)
**입력** 사용자로부터 온도 종류(섭씨 또는 화씨)와 온도 값을 입력 받기

**판단 및 계산**
  * `if kind == '섭씨':` 이면 화씨 공식(`temp * 1.8 + 32`)으로 계산
  * `else:` 이면 화씨로 간주하고 섭씨 공식(`(temp - 32) / 1.8`)으로 계산

**출력** 변환된 온도 결과 화면 표시

### 두 도형의 넓이 비교하기
**입력** 사각형(가로×세로)과 삼각형(밑변×높이/2)의 정보를 각각 입력 받아 넓이 계산

**로직**
  * `if 넓이1 > 넓이2:` 첫 번째 도형이 더 넓다고 출력
  * `elif 넓이1 < 넓이2:` 두 번째 도형이 더 넓다고 출력
  * `else:` 두 도형의 넓이가 같다고 출력
