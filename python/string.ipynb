# **문자열(String)**


## 문자열의 개요
0개 이상의 문자들이 모인 집합

(`'`)나 (`"`)로 감싸 표현함
* **문자열 연결** `+` 연산자 사용
* **문자열 반복** `*` 연산자 사용 ➡ 지정한 횟수만큼 문자열 반복 가능

---
**이스케이프(Escape) 문자**
* `\n` (줄 바꿈)
* `\t` (Tab 키)
* `\\` (\를 넣는 경우)
* `\'` ('를 넣는 경우)
* `\"` ("를 넣는 경우)

```python
s='''이름: 홍길동
나이: 25'''
print(s)
```

## 문자열 다루기
* **인덱싱(Indexing)** 문자열의 특정 위치에 있는 문자에 접근하는 방법 (**0번**부터 시작)
* **슬라이싱(Slicing)** 문자열의 일부분을 추출하는 방법

  `[시작:끝]` 형식에서 끝 인덱스 직전까지 추출
* **문자열 함수 및 메서드**
    * `len()` 문자열의 길이 반환
    * `upper()` / `lower()` 대문자/소문자로 변환
    * `replace()` 특정 문자를 다른 문자로 치환
    * `split()` 특정 구분자를 기준으로 문자열 분리 후 리스트 생성
    * `count()` 특정 문자의 출현 횟수를 셉니다.
    * `find()` / `index()` 특정 문자의 위치 탐색
    * `rstrip()` 문자열 오른쪽의 공백이나 특정 문자 제거
    * `lstrip()` 문자열 왼쪽의 공백이나 특정 문자 제거
    * `replace(기존 글자, 새 글자)` 문자열 변경
    * `str()` 문자열로 변환
    * `int()` 정수로 변환
    * `float()` 부동소수점수(실수)로 변환
    * `chr()` 정수에 해당하는 문자 반환
    * `ord()` 문자에 해당하는 정수 반환
    
    **`is~()`**
    
    뒷 내용을 만족시키는지 판별하는 메서드인 경우가 많음
    * `isalnum()` 알파벳과 숫자로만 구성
    * `islower()` 소문자로만 구성
    * `isupper()` 대문자로만 구성
    * `isspace()` 공백으로만 구성

```python
s = 'abc'
for i in range(len(s)):
  print(s[i], end='')
print('')
for i in s:
  print(i, end='')
```

```python
s = 'abcde'
print(s[1:3])
print(s[:3])
print(s[3:])
print(s[1:-2])
```

```python
while True:
  userId = input('아이디를 입력하세요: ')
  if len(userId) < 5 and userId.isalnum() == False:
    print('길이가 짧고(5자 이상), 사용자 아이디에는 알파벳과 숫자만 사용할 수 있습니다.')
  elif len(userId) < 5:
    print('길이가 짧습니다.(5자 이상)')
  elif userId.isalnum() == False:
      print('사용자 아이디에는 알파벳과 숫자만 사용할 수 있습니다.')
  else:
    print('아이디가 확인되었습니다!')
    break
```

```python
fav_lst = []
while True:
  slct_opt = input('버킷리스트 작성 / 버킷리스트 확인 / 종료: ')
  if slct_opt == '종료':
    break
  elif slct_opt == '버킷리스트 작성':
    input_fav = input(f'버킷리스트를 작성하세요(번지점프, 마라톤완주, 유럽여행): ')
    fav_lst.extend(input_fav.split(', '))
  elif slct_opt == '버킷리스트 확인':
    print("!--버킷리스트--!")
    for i in range(len(fav_lst)):
      print(f'{i+1}. {fav_lst[i]}', end=' ')
  else:
    print('제공하지 않는 기능입니다!')
```

### 모스 부호 번역기
**구조** 알파벳 리스트(`alphabet`)와 그에 대응하는 모스 부호 리스트(`morse`)를 생성합니다.

**로직**
* 문장 → 모스 부호
    * 입력받은 문장 한 글자씩 분리
    * `alphabet`에서 해당 글자의 인덱스 탐색
    * 동일한 인덱스를 가진 `morse`의 요소 연결 및 출력

* 모스 부호 → 문장
    * 입력받은 모스 부호를 공백 기준으로 `split()`
    * `morse` 리스트에서 위치를 찾아 해당하는 알파벳 조합


```python
alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
morse = ['.-','-...','-.-.','-..','.','..-.','--.','....','..','.---','-.-','.-..','--','-.','---','.--.','--.-','.-.','...','-','..-','...-','.--','-..-','-.--','--..']

input_name = input('당신의 이름은 영어(대문자)로 무엇입니까?: ')
for i in range(len(input_name)):
  if input_name[i] in alphabet:
    print(f'{morse[alphabet.index(input_name[i])]}', end=" ")
```

```python
alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
morse = ['.-','-...','-.-.','-..','.','..-.','--.','....','..','.---','-.-','.-..','--','-.','---','.--.','--.-','.-.','...','-','..-','...-','.--','-..-','-.--','--..']

input_morse = input('영어(대문자)로 변환할 모스부호(공백으로 구분)를 입력하세요: ')
morse_lst = input_morse.split(' ')
for i in range(len(morse_lst)):
  if morse_lst[i] in morse:
    print(alphabet[morse.index(morse_lst[i])], end="")
```
