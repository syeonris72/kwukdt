# **함수(Function)**

## 함수의 정의와 구성 요소
**함수** 코드의 재사용성을 높이는 가장 기본적인 단위

* 함수의 기본 선언 및 호출
  * **선언**
    * `def 함수명(매개변수, parameter):`
    * 매개변수(parameter)를 통해 값을 전달 받음

  * **호출**
    * `함수명(인자, argument)`

  * **재귀 함수**
  
  함수 내부에서 자기 자신을 호출하는 함수

* 변수의 유효 범위 (Scope)

  * **전역 변수 (Global)** 모듈(파일) 전체에서 접근 가능

  * **지역 변수 (Local)** 함수 내부에서 생성, 함수가 종료되면 실행 불가

  함수 안에서 전역 변수를 수정하려면 global 키워드를 사용해야 함
    
  But, 가급적 함수 인자로 전달하는 것이 유지보수에 좋음

## 함수의 확장 문법

* **가변 인자(`*args`, `**kwargs`)**

  입력 값의 개수가 정해지지 않았을 때 유용

* `*args`
  * 순차적인 입력에 대해서 개수에 상관없이 입력 받기 가능
  * 여러 개의 인자를 튜플(Tuple) 형태로 받음

* `**kwargs`
  * 키워드 인자에 대해서 어떤 이름의 인자 입력이든 자유롭게 입력 받기 가능
  * 키워드 인자(`key=value`)를 딕셔너리(Dict) 형태로 받음

* `func(*args, **kwargs)` 어떤 형태의 입력이든 가능

* **타입 힌트(Type Hint)**

  인자를 받을 때 타입을 지정
  ```
  def 함수(인자:타입):
  실행코드
  ```
* **람다(Lambda)**

  * 함수를 선언하는 문법
  * 한줄로 간단하게 정의되는 함수를 만들때 사용
  * 특정 입력 값을 고정한 새 함수를 만들때도 사용

  * `lambda x, y : x + y`

  * `sort()`, `filter()`, `map()` 등의 함수의 인자로 주로 사용

## 고급 함수형 프로그래밍
* 함수의 객체화
  * 함수를 변수에 대입하고 다른함수의 인자로 입력
  * 함수를 단순한 실행 단위가 아닌 **객체**로 다루는 기법

* 데코레이터(Decorator)

  * 기존 함수를 수정하지 않고 기능을 덧붙일 때 사용(로그 출력, 실행 시간 측정 등)
  * 함수의 기능을 변경, 확장할 수 있는 문법
  * 기존 함수의 기능은 수정하지 않고 함수 실행의 앞뒤에 미리 정의한 코드 실행 가능
  * 원본 함수를 입력받아, 기능을 추가한 새로운 함수를 만들어 반환하는 함수

  ```
  @데코레이터
  def 함수명(인자):
    실행코드
  ```
* 클로저(Closure)
  * 내부 함수가 외부 함수의 지역 변수를 기억하고 있는 상태로 함수 안 지역 변수를 함수 밖에서도 간접적으로 사용할 수 있게 하는 기술

  * 함수 안 지역 변수를 함수 밖에서도 간접적으로 사용 가능

  * 데이터를 은닉하거나 상태를 유지 가능

## 내장 함수 활용

| 함수 | 설명 | 예시 |
| --- | --- | --- |
| `zip()` | 여러 리스트의 요소를 짝지어 튜플로 반환 | `list(zip([1,2], ['a','b']))` |
| `enumerate()` | 인덱스와 요소를 동시에 반환 | `for i, v in enumerate(list):` |
| `isinstance()` | 객체가 특정 타입인지 확인 (`True`/`False`) | `isinstance(5, int)` |
| `all()` / `any()` | 모든 요소가 참인지 / 하나라도 참인지 확인 | `all([True, False]) -> False` |

* 내장 함수의 확장(`key` 사용)

    ```
    max(입력, key=함수)
    max(입력, key=lambda x: x.내장함수())
    ```
  * `max()` 등에서 `key` 인자를 사용하여 정렬 기준을 정의 가능


## 리스트와 문자열, 제어문 활용
* **문자열**

  * `format()` 확장

    * `{:,}` 1,000단위 콤마

    * `{:.2f}` 소수점 자리수 지정

    * `{:5}` 숫자 길이(공백 확보)

  * 아스키코드 기준 대문자 > 소문자 > 숫자(0-9) 순서로 사전식 비교가 이뤄짐

* **리스트**

* `" ".join(리스트)` 구분 기호로 리스트 결합

* `A if B else C` 삼항 연산자 (B가 참이면 A, 아니면 C)

* 리스트 컴프리헨션(List Comprehension)

  ```
  # [표현식 for 항목 in 반복가능객체 if 조건]
  evens = [x for x in range(10) if x % 2 == 0]
  ```
* 덧셈(+)으로 연결, 곱셈(*)으로 반복 가능

* **if문**
  * `all()`, `any()`를 조건식에 결합하여 사용 가능
  * 효율적인 조건식 (Short-circuit Evaluation)

    * **A or B** A가 참이면 전체를 참으로 판단

    * **A and B** A가 거짓이면 B를 확인하지 않고 바로 거짓으로 판단
    
  * **`True`/`False` 판별**

    단순 Boolean(True/False) 외에도 값이 있는지 없는지에 따라 참/거짓이 결정됨

    * **`False`의 의미**
      * `None`
      * 0 혹은 0.0
      * `""`(빈 문자열), `[]`(빈 리스트), `{}`(빈 딕셔너리)

    * **`True`의 의미** 위를 제외한 내용물이 있는 모든 객체

* **반복문**
  * 중첩 반복문
    * for 안에 for가 들어가는 구조
    * 내부에서 break, continue 사용 시 주의 필요

  * 반복문의 else
    * break 없이 반복문이 끝까지 돌았을 때 실행
    * 검색 알고리즘에서 대상을 찾지 못했을 경우 예외 처리를 할 때 유용함

  * **빅오 표기법(Big O Notation)**

    알고리즘의 성능을 입력 데이터 $n$의 크기에 따른 연산 횟수로 측정하는 방식

    * $O(1)$: 상수 시간 (데이터 양과 무관)
    * $O(n)$: 선형 시간 (데이터 양에 비례)
    * $O(n^2)$: 이차 시간 (이중 반복문)
    * $O(2^n)$: 지수 시간 (재귀 함수 피보나치 등, 매우 느림)
