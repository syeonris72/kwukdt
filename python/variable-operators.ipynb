# **변수(Variable) & 연산(Operators)**



## 변수 (Variable)

* 프로그램 실행 중 값을 저장하는 메모리 공간
* 변수에는 하나의 값만 저장(1:1)되며, 새로운 값을 넣으면 기존 값은 삭제됨

**대입 연산자 (`=`)** 오른쪽의 값을 왼쪽 변수에 저장
* `age = 20` (age라는 변수에 20을 대입)
---
## 변수 글자 규칙
* **첫 번째 글자** 문자 또는 밑줄(`_`)
* **나머지 글자** 문자, 숫자, 밑줄(`_`)
* 영문 대문자와 영문 소문자 구분
* 키워드(예약어)는 변수 이름 불가

```python
a1, b1, c1 = 7, 11, 24
print(a1, b1, c1)
```

```python
a2 = b2 = c2 = 10
print(a2, b2, c2)
```

## 자료형 (Data Type)

데이터의 종류에 따라 적절한 자료형을 사용해야 합니다.

* **정수형 (int)** 소수점이 없는 숫자 (10, -5)
* **실수형 (float)** 소수점이 있는 숫자 (3.14, 0.5)
* **문자열 (str)** 따옴표(`'`, `"`)로 감싸진 텍스트 ("Hello", 'Python')
* **불형 (bool)** 참(`True`) 또는 거짓(`False`)
---
## 진수 표시
| 진수 | 표시 |
| --- | --- |
| 8진수 | `0o(8진수 표시 숫자)` |
| 16진수 | `0x(16진수 표시 숫자)` |
| 2진수 | `0b(2진수 표시 숫자)` |

```python
print(type(2026))
print(type(3.14))
print(type('Python'))
print(type(True))
```

```python
num4 = 0o12
num5 = 0xA5
num6 = 0b1010
print(num4, num5, num6)
```

```python
num1 = 3.14
num2 = 1.23e5
num3 = 123000e-5
print(num1, num2, num3)
```

## 입력과 출력

* **출력 (`print`)** 변수나 값 화면에 표시
`print('age')` 문자열 출력, `print(age)` 변수 값 출력

* **입력 (`input`)** 사용자로부터 데이터를 입력 받음

  `input()`으로 받은 데이터 = 기본적으로 **문자열** 처리 (계산 필요 시 `int()`나 `float()`로 변환)
---
## f-string(f-문자열)
문자열 앞에 `f` 또는 `F`를 붙이고 문자열 안의 변수를 중괄호(`{}`)로 감싸서 표현하는 것
* 변수 값이 출력되는 형식 지정 가능
* 변수명 앞에 콜론(`:`)을 넣은 후 형식 지정
* 중괄호 안의 `d`는 정수, `f`는 부동소수점수 의미

**참고** 줄 바꿈이 필요하다면 `end = ""` 작성

```python
name = input('이름을 입력해주세요!')
age = input('나이를 입력해주세요!')
print(f'{name}, {age}세')
```

```python
print('반갑습니다')
score = int(input('점수를 입력해주세요!'))
print(f'당신의 점수는 {score}점 입니다.')
```

```python
# 원의 넓이 구하는 식
pi = 3.14159265
radius = int(input('원의 반지름: '))
area = pi * radius ** 2
print(f'반지름이 {radius}인 원의 넓이는 {area:.3f}')
```

## 연산자 (Operators)

산술 연산자를 통해 데이터 계산

| 연산자 | 기능 | 예시 (a=10, b=3) | 결과 |
| --- | --- | --- | --- |
| `+` | 더하기 | `a + b` | 13 |
| `-` | 빼기 | `a - b` | 7 |
| `*` | 곱하기 | `a * b` | 30 |
| `/` | 나누기 (실수) | `a / b` | 3.333... |
| `//` | **몫** | `a // b` | 3 |
| `%` | **나머지** | `a % b` | 1 |
| `**` | 거듭제곱 | `a ** b` | 1000 |
---
## 복합 대입 연산자
연산한 값을 다시 대입하는 연산자

| 연산자 | 기능 | 예시 (a=10, b=3) | `print(a)`의 결과 |
| --- | --- | --- | --- |
| `+=` | 더하기 | `a += b` | 13 |
| `-=` | 빼기 | `a -= b` | 7 |
| `*=` | 곱하기 | `a *= b` | 30 |
| `/=` | 나누기 (실수) | `a /= b` | 3.333... |
| `//=` | **몫** | `a //= b` | 3 |
| `%=` | **나머지** | `a %= b` | 1 |
| `**=` | 거듭제곱 | `a **= b` | 1000 |

```python
a3 = 0
a3 += 7
print(a3)

b3 = 5
b3 -= 3
print(b3)
```

```python
a4, b4 = 1, 2
a4 += b4
print(a4)
```

```python
# 문자열 연결 연산자 +
print('Hi' + 'five')
```

## **실습 예제**


### 시간 단위 변환기
**입력** 사용자로부터 초(second) 단위의 큰 숫자를 입력 받아 정수 변환

**로직**
* 전체 초를 3,600으로 나눈 **몫** = `hour` 저장
* `hour`을 60으로 나눈 **몫** = `minute`, **나머지** =  `second` 저장

**결과** 계산된 시간, 분, 초를 조합하여 화면 출력


```python
time = int(input('초 단위의 시간을 입력해주세요!'))
hour = time // 3600
minute = (time % 3600) // 60
second = (time % 3600) % 60
print(f'{hour}시간 {minute}분 {second}초 입니다.')
```

### 주문 금액 계산기(1)
**입력** 상품 가격과 주문 수량을 각각 입력 받아 숫자로 변환

**로직** 상품 가격 × 수량 ➡ 총 주문 금액 계산

**결과** "총 주문 금액은 OOO원입니다"와 같은 형태로 출력

```python
productPrice = int(input('상품 가격:'))
orderQuantity = int(input('주문 수량:'))
totalPrice = productPrice * orderQuantity
print(f'총 주문 금액은 {totalPrice}원 입니다.')
```

### 주문 금액 계산기(2)
**입력** 상품 가격 출력 후 주문 수량을 입력 받아 숫자로 변환

**로직** 상품 가격 × 수량 ➡ 총 주문 금액 계산

**결과** "총 주문 금액은 OOO원입니다"와 같은 형태로 출력

```python
print('아메리카노 2000 | 라떼 3000 | 녹차 2500')
americano_order = int(input('아메리카노 주문 수: '))
latte_order = int(input('라떼 주문 수: '))
greentea_order = int(input('녹차 주문 수: '))
sum = americano_order * 2000 + latte_order * 3000 + greentea_order * 2500
print(f'총 계산하실 금액은 {sum}원 입니다!')
```

```python
print('아메리카노 2000 | 라떼 3000 | 녹차 2500')
americano_price = 2000
latte_price = 3000
greentea_price = 2500
americano_order = int(input('아메리카노 주문 수: '))
latte_order = int(input('라떼 주문 수: '))
greentea_order = int(input('녹차 주문 수: '))
sum = americano_order * americano_price + latte_order * latte_price + greentea_order * greentea_price
print(f'총 계산하실 금액은 {sum}원 입니다!')
```
