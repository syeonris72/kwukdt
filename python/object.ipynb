# **객체(Object)**

## 객체와 클래스의 이해
* **객체 지향 프로그래밍(Object Oriented Programming)** 현실 세계의 모든 사물을 '객체'로 보고, 이를 프로그램으로 구현하는 기법
* **객체(Object)** 현실 세계에 존재하는 모든 것이 객체가 될 수 있으며, **속성**과 **메서드**로 구성
    * **속성(Attribute)** 객체의 상태나 특징을 나타내는 변수
    * **메서드(Method)** 객체의 동작이나 기능을 나타내는 함수
* **클래스(Class)**

  * 객체를 만들기 위한 '설계도' 또는 '틀'

  * 클래스를 통해 생성된 실제 사물을 '인스턴스(Instance)' 또는 '객체'라고 부름

## 클래스 생성 및 사용
* **클래스 정의** `class` 키워드를 사용하며, 내부에 속성(변수)과 메서드(함수) 정의
* **생성자 (`__init__`)** 객체가 생성될 때 자동으로 호출되는 특별한 메서드 (주로 객체의 초기 상태(속성) 설정에 사용)
* **`self` 키워드** 메서드 내부에서 객체 자신의 속성에 접근할 때 사용하며, 모든 메서드의 첫 번째 매개변수로 지정해야 함
* **정보 은닉(Private)** 속성 이름 앞에 `__`(언더바 두 개)를 붙여 클래스 외부에서 직접 수정하지 못하도록 보호할 수 있음

```python
class Student:
  def __init__(self, name, math, computer):
    self.__name = name
    self.__math = math
    self.__computer = computer
  def get_name(self):
    return self.__name
    return self.__computer
  def get_avg(self):
    return (self.__math + self.__computer) / 2
  def set_math(self, math):
    self.__math = math
  def set_computer(self, computer):
    self.__computer = computer

s1 = Student('kim', 96, 84)
s2 = Student('lee', 76, 92)
print(f'{s1.get_name()}의 수학과 컴퓨터 과목의 평균은 {s1.get_avg()}점 입니다.')
s1.set_computer(92)
print(f'{s1.get_name()}의 수학과 컴퓨터 과목의 평균이 {s1.get_avg()}점이 되었습니다.')
```

## 상속(Inheritance)
* 기존 클래스(부모 클래스)의 속성과 메서드를 그대로 물려받아 새로운 클래스(자식 클래스)를 만드는 기능
* 코드의 재사용성을 높여주며, 중복된 코드를 줄여 유지 보수가 쉬움
* **메서드 오버라이딩(Overriding)** 부모 클래스로부터 물려받은 메서드를 자식 클래스에서 용도에 맞게 새롭게 재정의하여 사용하는 것

```python
class Student:
  def __init__(self, name, computer):
    self.name = name
    self.computer = computer

  def get_name(self):
    return self.name

  def set_computer(self, computer):
    self.computer = computer

class Science(Student):
  def __init__(self, name, computer, science):
    super().__init__(name, computer)
    self.science = science

  def set_science(self, science):
    self.science = science

  def get_avg(self):
    return (self.science + self.computer) / 2

class Liberalarts(Student):
  def __init__(self, name, computer, social):
    super().__init__(name, computer)
    self.social = social

  def set_social(self, social):
    self.social = social

  def get_avg(self):
    return (self.social + self.computer) / 2

st1 = Science('lee', 92, 88)
print(st1.get_name())
print(st1.get_avg())
st1.set_computer(100)
print(st1.get_avg())
```

```python
class Figure:
  def __init__(self, area, perimeter):
    self.area = area
    self.perimeter = perimeter

  def get_area(self):
    return self.area

  def get_perimeter(self):
    return self.perimeter

class Circle(Figure):
  def __init__(self, radius):
    area = 3.14*radius**2
    perimeter = 3.14*radius*2
    super().__init__(area, perimeter)
    self.radius = radius

  def set_radius(self, radius):
    self.radius = radius
    area = 3.14*radius**2
    perimeter = 3.14*radius*2

  def get_radius(self):
    return self.radius

class Rectangle(Figure):
  def __init__(self, width=10, height=10): # 기본 값 지정
    area = width*height
    perimeter = 2*(width+height)
    super().__init__(area, perimeter)
    self.width = width
    self.height = height

  def set_width(self, width):
    self.width = width
    self.area = width*self.height
    self.perimeter = 2*(width+self.height)

  def set_hight(self, height):
    self.height = height
    self.area = self.width*height
    self.perimeter = 2*(self.width+height)

  def get_width(self):
    return self.width

  def get_height(self):
    return self.height

f1 = Circle(7)
print(f1.get_area())
print(f1.get_perimeter())

f2 = Rectangle(10, 60)
print(f2.get_area())
print(f2.get_perimeter())
```

### 자동차 클래스와 객체
**설계** `Car` 클래스 생성, 모델명(`__model`)과 연비(`__fuelefficiency`)를 속성으로 정의

**정보 보호** 속성들을 `private`으로 설정 ➡ 외부의 직접 접근을 막고, 안전하게 값을 읽거나 수정할 수 있는 전용 메서드(`get_model`, `set_fuelefficiency` 등)를 만들기

**로직** 주행 거리를 입력받아 (주행 거리 / 연비) 공식을 통해 필요한 연료량을 계산하여 반환하는 메서드 구현

**실행** 정의된 클래스를 바탕으로 `car1`, `car2` 등 서로 다른 모델명과 연비를 가진 실제 자동차 객체들을 만들어 실행

```python
class Car:
  def __init__(self, model, fuelefficiency):
    self.__model = model
    self.__fuelefficiency = fuelefficiency
  def get_model(self):
    return self.__model
  def set_fuelefficiency(self, fuelefficiency):
    self.__fuelefficiency = fuelefficiency
  def get_fuelefficiency(self):
    return self.__fuelefficiency
  def get_fuel(self, distance):
    return distance / self.__fuelefficiency
car1 = Car('소나타', 10)
car2 = Car('아반떼', 15)
print(f'{car1.get_model()}의 연료량은 {car1.get_fuel(100)} 입니다.') # () 안에 거리 입력
print(f'{car2.get_model()}의 연료량은 {car2.get_fuel(100)} 입니다.')
```
