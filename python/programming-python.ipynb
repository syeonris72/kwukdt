# **프로그래밍(Programming) & 파이썬(Python)**



## 프로그래밍 언어
* **프로그래밍의 이해**:
    * **프로그램(Program)** 컴퓨터로 문제를 해결하기 위해 작성하는 명령어의 모음
    * **프로그래밍(Programming)** 프로그램을 작성하는 과정
    * **프로그래머(Programmer)** 프로그램을 작성하는 사람
* **프로그래밍 언어의 종류**
    * **저급 언어** 기계 중심의 언어, 하드웨어 지향적 (어셈블리어)
    * **고급 언어** 사람이 이해하기 쉬운 언어와 기호 사용 (C, 자바, 파이썬 등)
    
    **언어 번역** 고급 언어 ➡ 기계어 변환 과정 필요
* **라이브러리(Library)**
  프로그램에서 자주 사용되는 부분 프로그램들을 모아 놓은 것
    * **넘파이(NumPy)**: 벡터 및 행렬 연산
    * **판다스(Pandas)**: 데이터 처리 및 분석
    * **사이킷런(Scikit-learn)**: 머신러닝
    * **케라스(Keras)**: 딥러닝
    * **텐서플로(TensorFlow)**: 구글 라이브러리, 딥러닝과 머신러닝
    * **파이토치(PyTorch)**

```python
print('파이썬 수업 시작!')
```

## 파이썬의 개념
* **파이썬(Python)** 1991년 귀도 반 로섬이 발표한 프로그래밍 언어
* **주요 특징**
    * 문법 간결, 사람의 언어와 비슷, 배우기 쉬움
    * 다양한 분야(데이터 분석, 인공지능, 웹 개발 등)에서 활용 가능한 라이브러리 제공
    * 무료 사용 가능한 오픈 소스

## 파이썬 개발 환경
* **구글 코랩(Google Colab)** 웹 브라우저에서 파이썬 코드를 작성하고 실행할 수 있는 서비스
    * 별도의 설치 없이 무료 사용 가능
    * 작성한 파일은 구글 드라이브(`Colab Notebooks` 폴더)에 자동 저장
* **구성 요소**
    * **코드 셀** 파이썬 코드를 입력 및 실행하는 공간
    * **텍스트 셀** 설명이나 메모를 마크다운 형식으로 작성하는 공간

## 처음 만나는 프로그램
* **프로그램 실행** 코드 셀에 소스 코드를 입력한 후 실행 버튼을 누르거나 단축키(`Ctrl + Enter`)를 사용하여 결과 확인
* **파일 관리** 파일 이름 변경 또는 구글 드라이브를 통해 저장된 파일 불러오기 가능
