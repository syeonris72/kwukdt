# **클래스(Class)**

## 클래스의 기본 개념 및 구성 요소
* **클래스** 객체를 만들기 위한 설계도

* **인스턴스** 설계도로 만든 실체
* 클래스의 선언 및 인스턴스 생성
  ```
  class Starbucks:
      brand = "Starbucks"  # 클래스 변수

      def __init__(self, name, price): # 생성자
          self.name = name           # 인스턴스 변수
          self.price = price         # 인스턴스 변수

      def order(self):               # 메서드
          print(f"{self.brand}에서 {self.name}을(를) 주문합니다.")

  # 인스턴스 생성
  americano = Starbucks("아메리카노", 4500)
  americano.order()
  ```
* 주요 용어
  * **인스턴스(Instance)** 클래스라는 설계도를 바탕으로 메모리에 실제 구현된 객체

  * **메서드(Method)** 클래스 내부의 함수로, 객체의 '행동'을 정의

  * **애트리뷰트(Attribute)** 객체의 상태나 성질. 인스턴스 변수와 클래스 변수를 통칭

  * **`self`** 객체 자기 자신을 가리키는 키워드로 메서드 호출 시 어떤 객체가 호출했는지 구분함

  * **생성자(Constructor)** `__init__` 메서드로 객체 생성 시 초기값을 설정하기 위해 자동 호출됨

## 변수의 종류
변수가 선언된 위치에 따라 공유 범위가 달라짐

* **인스턴스 변수(Instance Variable)** 각 인스턴스가 개별적으로 가지는 변수(`self.변수명`)

* **클래스 변수(Class Variable)** 모든 인스턴스가 공유하는 변수(클래스 내부, 메서드 외부 정의)

````python
class Korean:
  country = 'korea'
  def set_name(self, name):
    self.name = name

k1 = Korean()
k2 = Korean()

k1.set_name('kim')
k2.set_name('lee')

print(k1.name, k1.country)
print(k2.name, k2.country)
````

## 매직 메서드(Magic Methods)
이름 앞뒤에 언더바 두 개(__)가 붙은 특수 메서드로, 특정 상황에서 자동으로 호출됨

| 메서드 | 역할 | 특이사항 |
| --- | --- | --- |
| `__init__()` | **생성자** 인스턴스 초기화 수행 | `return` 사용 불가 |
| `__call__()` | 인스턴스를 **함수처럼 호출**할 때 실행 | `obj()` 형태 |
| `__len__()` | `len(obj)` 호출 시 실행 | `self` 외 인자 불가 |
| `__getitem__()` | **인덱싱/슬라이싱** `(obj[0])` 수행 시 실행 | 대괄호 안의 정보가 인자로 전달 |
| `__add__()` | **덧셈 연산자(+)** 사용 시 호출 | 객체 간의 더하기 정의 |



````python
class Multiplier:
  def __init__(self, mul):
    self.mul = mul
  def __call__(self, num):
    return self.mul * num

triple = Multiplier(3)
double = Multiplier(2)

print(triple(10))
print(double(10))
````

````python
class Data:
  def __init__(self, data):
    self.data = [-1] + data
  def __len__(self):
    return len(self.data)

d = Data([10, 20, 30])
print(len(d))
````

## 상속(Inheritance)과 오버라이딩(Overriding)
**상속** 기존 클래스(부모)의 기능을 물려받아 새로운 클래스(자식)를 만드는 것
* 코드 재사용 및 기능 확장.

  ```
  class Parent:
      def set_name(self, name):
          self.name = name

  class Child(Parent):
      def __init__(self, name):
          super().set_name(name) # 부모의 메서드 활용
  ```
**오버라이딩(Overriding)** 부모의 메서드를 자식 클래스에서 입맛에 맞게 재정의(덮어쓰기)하는 것
* `super()` 부모 클래스의 메서드를 호출할 때 사용


````python
class FindPath:
  def find(self, start, dest):
    print(start, dest)

class FPBicycle(FindPath):
  def get_bicycle_road(self):
    print('get bicycle road')

bicycle = FPBicycle()
bicycle.find('Seoul', 'Busan')
````

````python
class FindPath:
  def find(self, start, dest):
    print(start, dest)

class FPBicycle(FindPath):
  def find(self, start, dest):
    print('bicycle:', start, dest)

bicycle = FPBicycle()
bicycle.find('Seoul', 'Busan')
````

````python
class FindPath:
  def find(self, start, dest):
    print(start, dest)

class FPBicycle(FindPath):
  def find(self, start, dest):
    super().find(start, dest)
    print('withBicycle')

bicycle = FPBicycle()
bicycle.find('Seoul', 'Busan')
````

## 데코레이터를 이용한 메서드 확장
* **`@property`(Getter & Setter)**
  * 인스턴스 변수에 접근하거나 값을 할당할 때 특정 로직(실행 코드)을 거치게 함
  ```
  class Member:
      @property
      def age(self): return self._age
      
      @age.setter
      def age(self, value): self._age = value
  ```
* **`@staticmethod`(정적 메서드)**
  * 인스턴스나 클래스 변수에 접근하지 않고 독립적으로 동작하는 메서드(인스턴스나 클래스 각각 호출 가능)
  * `self`가 없으면 정적 메서드
  * 객체를 생성하지 않고도 호출 가능
  ```
  class 클래스명:
  @staticmethod
  def 메서드(인자):
    실행코드
  ```
* **`@classmethod`(클래스 메서드)**
  * 첫 번째 인자로 클래스(cls)를 받음
  * 클래스 변수를 읽거나 수정할 때 주로 사용함
  ```
  class 클래스명:
  @classmethod
  def 메서드(cls, 변수): # 첫번째 인자가 클래스를 가리키며 클래스변수에 접근하는데 사용
    실행 코드
  ```

## 동적 애트리뷰트 관리
파이썬은 실행 중에 객체의 속성을 가져오거나 설정하는 유연한 도구를 제공함
* 내장 함수(External)
  * **`getattr(obj, name[, default])`**
    * 문자열 이름을 이용해 속성값 획득
    * 존재하지 않으면 에러 혹은 기본값 리턴
    ```
    getattr(객체, 이름[, 기본값])

    class Member:
      def __init__(self, name):
        self.name = name

    m = Member('kim')
    print(getattr(m, 'name)) # m.name 리턴(존재하지 않으면 에러 발생, 기본값이 있을 경우 기본값 리턴)
    ```
  * **`setattr(obj, name, value)`**
    * 문자열 이름을 이용해 속성값 동적 할당
    ```
    setattr(객체, 이름, 값)

    class Member:
      def __init__(self, name):
        self,name = name
        
    m = Member('kim')
    setattr(m, 'name', 'lee') # setattr() 함수의 리턴값은 None 이다.
    ```
* 매직 메서드(Internal)
  * **`__getattr__(self, name)`**
    * 객체에 찾는 속성이 없을 때만 마지막 수단으로 호출되어 에러 방지
    ```
    class 클래스명:
    def __getattr__(self, name):
      실행코드
    ```
  * **`__setattr__(self, name, value)`**
    * 객체의 속성이 설정될 때마다 항상 호출되어 설정 방식 커스텀

## 내장 함수(External)와 매직 메서드(Internal)의 비교
* `getattr()` vs `__getattr__`
  * **`getattr(obj, name)`**
    * 객체의 속성 값을 가져오는 내장 함수(`obj.name`과 결과 같음)
    * 속성이 없으면 에러(AttributeError) 발생

  * **`__getattr__(self, name)`**
    * 객체에 찾는 속성이 없을 때만 마지막 수단으로 호출되는 매직 메서드
    * 에러가 나기 직전에 실행되어 에러를 막을 수 있음

  | 구분 | `getattr()` | `__getattr__` |
  | --- | --- | --- |
  | 성격 | 외부에서 값을 읽어오는 **도구** | 값이 없을 때를 대비한 방어 기제 |
  | 시점 | 개발자가 직접 코드 에서 호출할 때 | `obj.abc`를 찾았는데 `abc`가 메모리에 없을 때 **자동 실행** |
  | 특징 | 속성이 없으면 에러(AttributeError) 발생 | 에러가 나기 직전에 실행되어 에러를 막을 수 있음 |

* `setattr()` vs `__setattr__`
  * **`setattr(obj, name, value)`**
    * 객체의 속성을 설정하기 위해 사용하는 내장 함수(`obj.name = value`와 결과 같음)

  * **`__setattr__(self, name, value)`**
    * 객체의 속성이 설정될 때 내부적으로 자동 호출되는 매직 메서드
    * 속성 설정 방식을 커스텀(재정의)하고 싶을 때 사용

  | 구분 | `setattr()` | `__setattr__` |
  | --- | --- | --- |
  | 성격 | 외부에서 속성을 조작하는 **도구** |  클래스 내부에 정의된 **동작 규칙** |
  | 시점 | 속성 이름을 문자열로 가지고 있을 때 | 속성이 저장되는 방식을 통제하고 싶을 때 |
  | 작동 방식 | 내부적으로 결국 `__setattr__`을 실행함 | 모든 속성 설정(`obj.a = 1`) 시 호출됨 |
