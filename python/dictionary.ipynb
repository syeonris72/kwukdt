# **딕셔너리(Dictionary)**

## 딕셔너리의 개요
* 리스트와 유사한 여러 개의 데이터를 저장하는 자료형
* 키(Key)와 값(Value)의 쌍으로 이루어짐 (`{'키 key' : 값 value}`)
  * **키(key)**
    * 변경 불가능한 자료형만 사용 가능 (리스트, 딕셔너리, 집합 사용 불가)
    * **키의 조건** 딕셔너리 내에서 키는 중복될 수 없으며 유일해야 함 (만약 동일한 키가 여러 개 있다면 마지막 항목만 유효)
  * **값(Value)** 어떤 자료형이든 사용 가능
* 중괄호 `{ }` 사용, 각 항목은 `키:값` 형식으로 작성하고 쉼표(`,`)로 구분
* non sequence type으로 순서 X
* **리스트 vs 딕셔너리**

  리스트: 인덱스(번호)를 사용하여 항목에 접근
  
  딕셔너리: 사용자가 정의한 '키'를 사용하여 데이터(값)에 접근



## 딕셔너리 데이터 조회/추가/삭제 방법

* **접근** 딕셔너리 이름 뒤 대괄호 `[ ]` 안에 키 이름을 넣어 값을 가져오기
* **항목 추가 및 변경** 새로운 키에 값을 대입하면 항목이 추가되고, 기존에 있는 키에 값을 대입하면 해당 값이 수정됨
* `del` 특정 키를 지정하여 해당 항목(키와 값의 쌍) 삭제
* `clear` 딕셔너리 내부의 모든 데이터를 한꺼번에 비우기

## 딕셔너리 전용 함수

* `keys()` 딕셔너리에 포함된 모든 키들을 모아서 보여줌
* `values()` 딕셔너리에 포함된 모든 값들을 모아서 보여줌
* `items()` 키와 값의 쌍을 튜플 형태로 묶어서 반환, 반복문과 함께 자주 사용
* `in` 특정 키가 딕셔너리에 들어있는지 확인하여 참/거짓 반환
* `get()` 주어진 키(key)에 해당하는 값을 반환
* `pop()` 맨 마지막 요소를 제거하고, 그 제거된 요소를 반환

## **실습 예제**


### 프로그래밍 언어 선호도 조사
**입력** 사용자로부터 사용 가능한 프로그래밍 언어를 여러 번 입력 받기

**로직**
  * 입력받은 언어가 이미 딕셔너리의 키로 존재하면, 해당 키의 값을 1 증가
  * 존재하지 않는 새로운 언어라면, 해당 언어를 키로 추가하고 값에 1 저장

**출력** 각 언어별로 몇 번씩 입력되었는지 집계된 데이터 출력

### 영어 단어 퀴즈 (동물 이름)
**데이터 생성** 한국어 동물 이름을 키로, 영어 단어를 값으로 하는 딕셔너리 생성

**실행** `items()` 함수를 사용하여 딕셔너리의 모든 항목 순회

**입력** 한국어 이름을 제시하고 사용자에게 영어 단어 입력받기

**출력** 사용자가 입력한 단어와 딕셔너리의 '값(영어 단어)'이 일치하는지 비교하여 정답 여부 출력
