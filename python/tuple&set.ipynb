# **튜플(Tuple)과 집합(Set)**

## 튜플 (Tuple)

* `(a, b)` 또는 괄호 없이 콤마로 구분
* 파이썬 내부 동작에서 주로 사용, 쉼표를 튜플 생성자로 사용하니 괄호는 생략 가능
* `(a, )` 하나의 요소만 갖는 튜플 선언시 요소 뒤에 콤마
* 리스트와 비슷하지만 튜플은 값의 수정 불가능, 불변
* 튜플 자체는 수정이 불가능하지만, 원소로 수정 가능한 리스트나 딕셔너리를 담는 것은 가능


## 집합 (Set)
* 집합의 원소는 반드시 **값이 변하지 않는(Hashable)** 상태 (리스트나 딕셔너리는 언제든 값이 바뀔 수 있기 때문에 집합의 원소가 될 수 없음)
* 중복되지 않은 요소들로 구성된 자료형
* 집합 내의 모든 요소들은 순서 X
* 인덱싱이나 슬라이싱으로 집합 내의 요소에 접근 X
```
set1 = {1,2,3}
set1 = {} # 딕셔너리로 인식
set2 = set()
```
* **연산자**
  * `|` 합집합
  * `&` 교집합
  * `-` 차집합
  * `^` 대칭 차집합(합집합에서 교집합을 뺀 집합)
* **내장함수**
  * `add()` 요소 추가
  * `remove()` 요소 삭제
  * `update()` 요소 수정 및 추가


## 자료형 및 데이터 타입 정리

`None` 값이 없는 상태를 나타내는 자료형

**복소수(complex number)**
* 실수부 + 허수부j
* `complex(실수부, 허수부)`

**Bytes 자료형**
* 주로 네트워크 통신, 파일 입출력(이미지, 실행 파일), 암호화 등 바이너리 데이터를 다룰 때 사용
* 변경 불가능(Immutable)
* 0부터 255 사이의 정수(8비트)만 원소로 가능
* `b''` 형태로 사용

### **데이터 타입**

* 숫자 데이터 타입(numeric type): int, float, complex

* 시퀀스 자료형(sequence type): list, tuple, range

* 문자 시퀀스 자료형(text sequence type): str

* 집합 자료형(set types): set

* 매핑 자료형(mapping types): dict

* 기타: none, boolean, functions
