```python
print("tickets 데이터프레임 (일부):")
display(tickets.head())

print("\nclass_info 데이터프레임:")
display(class_info)

# inner join
merged_inner = pd.merge(tickets, class_info, on='pclass', how='inner')
print("\n1. Inner Join (공통 pclass):")
display(merged_inner.head())

# left join
merged_left = pd.merge(tickets, class_info, on='pclass', how='left')
print("\n2. Left Join (tickets 기준):")
display(merged_left.head())

# right join
merged_right = pd.merge(tickets, class_info, on='pclass', how='right')
print("\n3. Right Join (class_info 기준):")
display(merged_right)

# outer join
merged_outer = pd.merge(tickets, class_info, on='pclass', how='outer')
print("\n4. Outer Join (모든 pclass):")
display(merged_outer.head())
```

```python
df_passenger_id_diff = passengers[['passenger_id', 'Name', 'Sex']].copy()
df_passenger_id_diff.columns = ['p_id', 'Name', 'Sex']
df_tickets_id_diff = tickets[['passenger_id', 'fare', 'embarked']].copy()

print("df_passenger_id_diff (p_id 열 사용):")
display(df_passenger_id_diff.head())

print("\ndf_tickets_id_diff (passenger_id 열 사용):")
display(df_tickets_id_diff.head())

merged_diff_cols = pd.merge(
    df_passenger_id_diff, df_tickets_id_diff,
    left_on='p_id', right_on='passenger_id',
    how='left'
)

print("\n다른 이름의 열(p_id, passenger_id)을 기준으로 병합 (Left Merge):")
display(merged_diff_cols.head())

merged_diff_cols_cleaned = merged_diff_cols.drop(columns=['passenger_id'])
print("\n병합 후 중복 키 열 제거:")
display(merged_diff_cols_cleaned.head())
```

```python
df_left_idx = pd.DataFrame({'value_l': ['A', 'B', 'C']},
                             index=[0, 1, 2])
df_right_idx = pd.DataFrame({'value_r': [10, 20, 30]},
                              index=[2, 0, 1]) # 인덱스 순서가 다름

print("왼쪽 데이터프레임 (인덱스 기준):")
display(df_left_idx)

print("\n오른쪽 데이터프레임 (인덱스 기준):")
display(df_right_idx)

merged_by_index = pd.merge(
    df_left_idx, df_right_idx,
    left_index=True, right_index=True,
    how='inner'
)

print("\n인덱스를 기준으로 병합된 데이터프레임 (Inner Merge):")
display(merged_by_index)
```

```python
print("passengers 데이터프레임 (일부):")
display(passengers[['passenger_id', 'Name']].head())

print("\ntickets 데이터프레임 (일부):")
display(tickets[['passenger_id', 'fare', 'pclass']].head())

# 1:1 병합
merged_1_to_1 = pd.merge(passengers[['passenger_id', 'Name']],
                         tickets[['passenger_id', 'fare', 'pclass']],
                         on='passenger_id',
                         how='inner')

print("\n1:1 병합 결과 (inner join):")
display(merged_1_to_1.head())
print(f"원래 passengers 행 수: {len(passengers)}")
print(f"원래 tickets 행 수: {len(tickets)}")
print(f"병합된 데이터프레임 행 수: {len(merged_1_to_1)}")

assert merged_1_to_1['passenger_id'].is_unique, "passenger_id가 고유하지 않습니다!"
```

```python
print("tickets 데이터프레임 (pclass 분포):")
display(tickets['pclass'].value_counts().sort_index())

print("\nclass_info 데이터프레임:")
display(class_info)

# n:1 병합
merged_n_to_1 = pd.merge(tickets[['passenger_id', 'pclass', 'fare']],
                         class_info[['pclass', 'class_name', 'base_fare']],
                         on='pclass',
                         how='inner')

print("\nn:1 병합 결과 (inner join):")
display(merged_n_to_1.head())

print(f"원래 tickets 행 수: {len(tickets)}")
print(f"병합된 데이터프레임 행 수: {len(merged_n_to_1)}")
assert len(merged_n_to_1) == len(tickets), "병합 후 행 수가 예상과 다릅니다!"
```

```python
df_products = pd.DataFrame({
    'product_id': [1, 2, 3, 1, 2],
    'product_name': ['Apple', 'Banana', 'Cherry', 'Apple', 'Banana'],
    'category_code': ['F', 'F', 'F', 'S', 'D'] # F:과일, S:간식, D:음료
})

df_categories = pd.DataFrame({
    'category_code': ['F', 'F', 'S', 'D', 'D'],
    'feature': ['Sweet', 'Healthy', 'Crunchy', 'Cold', 'Fizzy']
})

print("df_products (category_code 중복):")
display(df_products)

print("\ndf_categories (category_code 중복):")
display(df_categories)

# n:n 병합
merged_n_to_n = pd.merge(df_products, df_categories, on='category_code', how='inner')

print("\nn:n 병합 결과 (inner join):")
display(merged_n_to_n)

print(f"원래 df_products 행 수: {len(df_products)}")
print(f"원래 df_categories 행 수: {len(df_categories)}")
print(f"병합된 데이터프레임 행 수: {len(merged_n_to_n)}")

assert len(merged_n_to_n) == 9, f"n:n 병합 후 행 수가 예상과 다릅니다! 예상: 9, 실제: {len(merged_n_to_n)}"
```
