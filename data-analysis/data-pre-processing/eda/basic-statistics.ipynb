```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df
```

```python
print(df.shape) # (행, 열)
```

```python
# 결측치 확인
print(df.isnull().sum())
```

```python
print("\n전체 칼럼 평균:\n", df.mean(numeric_only=True).round(2))
print("\n전체 칼럼 중앙값:\n", df.median(numeric_only=True).round(2))
print("\n전체 칼럼 최댓값:\n", df.max(numeric_only=True).round(2))
print("\n전체 칼럼 최솟값:\n", df.min(numeric_only=True).round(2))
print("\n전체 칼럼 25% 분위수:\n", df.quantile(0.25, numeric_only=True).round(2))
print("\n전체 칼럼 75% 분위수:\n", df.quantile(0.75, numeric_only=True).round(2))
print("\n전체 칼럼 분산:\n", df.var(numeric_only=True).round(2))
print("\n전체 칼럼 표준편차:\n", df.std(numeric_only=True).round(2))
print("\n전체 칼럼 최빈값:\n", df.mode(numeric_only=True))
```

```python
# df.describe()를 사용하여 데이터프레임의 기초 통계량 요약
print("데이터프레임의 기초 통계량 요약 (df.describe()):\n", df.describe())
```
