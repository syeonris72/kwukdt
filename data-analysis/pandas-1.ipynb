```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame([
    ["김", 80, 20 , "여성"],
    ["이", 92, 22 , "여성"],
    ["박", 76, 27 , "남성"],
    ["최", 88, 39 , "여성"],
    ["장", 64, 14 , "남성"]],
    columns=["이름", "점수", "나이", "성별"]
)
```

```python
display(df)
```

```python
print("DataFrame 구성:")
print(df.shape)
print("\nDataFrame 열:")
print(df.columns)
print("\nDataFrame 데이터 유형:")
print(df.dtypes)
print("\nDataFrame 정보:")
df.info()
print("\n처음 5개 행:")
display(df.head())
print("\n마지막 5개 행:")
display(df.tail())
```

```python
print("'이름' 열을 Series로 추출:")
display(df['이름'])

print("\n'이름'과 '점수' 열을 DataFrame으로 추출:")
display(df[['이름', '점수']])
```

```python
print("df.loc[]를 사용하여 인덱스 0인 행 추출:")
display(df.loc[0])

print("\ndf.loc[]를 사용하여 인덱스 1, 3, 4인 행 추출:")
display(df.loc[[1, 3, 4]])
```

```python
print("df.iloc[0]를 사용하여 첫 번째 행 추출:")
display(df.iloc[0])

print("\ndf.iloc[[1, 3, 4]]를 사용하여 위치 1, 3, 4에 있는 행 추출:")
display(df.iloc[[1, 3, 4]])

print("\ndf.iloc[-1]를 사용하여 마지막 행 추출:")
display(df.iloc[-1])
```

```python
print("df.loc[]를 사용하여 0, 1, 2 행과 '이름', '점수' 열 추출:")
display(df.loc[[0, 1, 2], ['이름', '점수']])

print("\ndf.iloc[]를 사용하여 위치 0, 1, 2 행과 위치 0, 1 열 추출:")
display(df.iloc[[0, 1, 2], [0, 1]])
```

```python
print("df.iloc[] 열 슬라이싱을 사용하여 '점수' 및 '나이' 열 추출:")
display(df.iloc[:, 1:3])
```

```python
print("groupby를 사용하여 '성별' 별 '점수' 평균:")
display(df.groupby('성별')['점수'].mean())
```

```python
print("groupby와 reset_index()를 사용하여 '성별' 및 '나이' 별 '점수' 평균:")
display(df.groupby(['성별', '나이'])['점수'].mean().reset_index())
```

```python
print("성별별 고유한 '나이' 수 (nunique() 사용):")
display(df.groupby('성별')['나이'].nunique())

print("\n각 '성별'의 빈도수 (value_counts() 사용):")
display(df['성별'].value_counts())
```
