```python
# 넘파이(Numpy)와 판다스(Pandas) 라이브러리 불러오기
import numpy as np
import pandas as pd
```

```python
# 데이터 프레임(DataFrame) 만들기
df = pd.DataFrame([
    ["김", 80, 20 , "여성"],
    ["이", 92, 22 , "여성"],
    ["박", 76, 27 , "남성"],
    ["최", 88, 39 , "여성"],
    ["장", 64, 14 , "남성"]],
    columns=["이름", "점수", "나이", "성별"]
)
```

```python
display(df)
```

```python
# 데이터 프레임(DataFrame) 정보 확인
print(f"DataFrame 구성: {df.shape}")
print(f"\nDataFrame 열: {df.columns}")
print(f"\n===== DataFrame 데이터 유형 =====\n{df.dtypes}")
print(f"\n===== DataFrame 정보 =====")
df.info()
print(f"\n===== 처음 5개 행 =====")
display(df.head())
print(f"\n===== 마지막 5개 행 =====")
display(df.tail())
```

```python
# '이름' 열을 시리즈(Series)로 추출"
display(df['이름'])

# '이름'과 '점수' 열을 데이터 프레임(DataFrame)으로 추출"
display(df[['이름', '점수']])
```

```python
# df.loc[]를 사용하여 인덱스 0인 행 추출
display(df.loc[0])

# df.loc[]를 사용하여 인덱스 1, 3, 4인 행 추출
display(df.loc[[1, 3, 4]])
```

```python
# df.iloc[0]를 사용하여 첫 번째 행 추출
display(df.iloc[0])

# df.iloc[[1, 3, 4]]를 사용하여 위치 1, 3, 4에 있는 행 추출
display(df.iloc[[1, 3, 4]])

# df.iloc[-1]를 사용하여 마지막 행 추출
display(df.iloc[-1])
```

```python
# df.loc[]를 사용하여 0, 1, 2 행과 '이름', '점수' 열 추출
display(df.loc[[0, 1, 2], ['이름', '점수']])

# df.iloc[]를 사용하여 위치 0, 1, 2 행과 위치 0, 1 열 추출
display(df.iloc[[0, 1, 2], [0, 1]])
```

```python
# df.iloc[] 열 슬라이싱을 사용하여 '점수' 및 '나이' 열 추출
display(df.iloc[:, 1:3])
```

```python
# groupby를 사용하여 '성별' 별 '점수' 평균 (소수점 둘째자리 반올림)
display(df.groupby('성별')['점수'].mean().round(2))
```

```python
# groupby와 reset_index()를 사용하여 '성별' 및 '나이' 별 '점수' 평균 (소수점 둘째자리 반올림)
avg_df = df.groupby(['성별', '나이'])['점수'].mean().reset_index().round(2)
display(avg_df)
```

```python
# 성별별 고유한 '나이' 수 (nunique() 사용)
display(df.groupby('성별')['나이'].nunique())

# 각 '성별'의 빈도수 (value_counts() 사용)
display(df['성별'].value_counts())
```
