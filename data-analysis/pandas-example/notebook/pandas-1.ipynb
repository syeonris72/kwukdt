```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/kwukdt/data-analysis/data-collection/gapminder.tsv', sep='\t')
```

```python
# 1. 데이터 구조 파악하기

# df.shape, df.columns, df.dtypes
print(f"(행, 열): {df.shape}")
print(f"열 이름 목록: {df.columns}")
print(f"\n--- 각 열의 자료형 ---\n{df.dtypes}")

# df.info()
print(f"\n--- 데이터 프레임 종합 정보 ---")
df.info()

# df.head()
print(f"\n--- 상위 5개 열 ---")
display(df.head())

# df.tail()
print(f"\n--- 하위 5개 열 ---")
display(df.tail())
```

```python
# 2. 데이터 추출 - 열(Column) 데이터 추출

# 시리즈(Series) 추출
country_df = df['country']
print("country_df의 자료형:", type(country_df))
display(country_df.head())

# 데이터 프레임(DataFrame) 추출 (리스트 활용)
subset = df[['country', 'continent', 'year']]
print("\nsubset의 자료형:", type(subset))
display(subset.head())
```

```python
# 3. 데이터 추출 - 행(Row) 데이터 추출

# loc (Label-based) - 인덱스 값으로 추출
print("--- 인덱스 라벨이 0인 단일 행 ---")
display(df.loc[0])
print("\n--- 인덱스 라벨이 99인 단일 행 ---")
display(df.loc[99])
print("\n--- 인덱스 라벨이 0, 99, 999인 여러 행 ---")
display(df.loc[[0, 99, 999]])

# iloc (Integer-based) - 위치 번호로 추출
print("\n--- 가장 첫 번째 위치의 행(0번 위치) ---")
display(df.iloc[0])
print("\n--- 100번째 위치의 행(99번 위치) ---")
display(df.iloc[99])
print("\n--- 가장 마지막 위치의 행 ---")
display(df.iloc[-1])
print("\n--- 0, 99, 999번째 위치의 여러 행 ---")
display(df.iloc[[0, 99, 999]])
```

```python
# 4. 데이터 추출 - 행(Row)과 열(Column) 데이터 추출

# loc(Label-based)
print(f"인덱스 라벨이 42인 행의 'country' 열에 해당하는 값: {df.loc[42, 'country']}")
print("\n--- 인덱스 라벨이 0, 99, 999인 행들과 'country', 'lifeExp', 'gdpPercap' 열들에 해당하는 데이터 ---")
display(df.loc[[0, 99, 999], ['country', 'lifeExp', 'gdpPercap']])

# iloc(Integer-based)
print(f"\n42번째 위치의 행과 0번째 위치의 열에 해당하는 값: {df.iloc[42, 0]}")
print("\n--- 0, 99, 999번째 위치의 행들과 0, 3, 5번째 위치의 열들에 해당하는 데이터 ---")
display(df.iloc[[0, 99, 999], [0, 3, 5]])
```

```python
# 5. 데이터 추출 - 슬라이싱(Slicing)과 범위 지정

# iloc 슬라이싱
print("--- 모든 행에서 처음 3개 열 추출 (0, 1, 2번) ---")
display(df.iloc[:, :3].head())

# iloc 슬라이싱
print("\n--- 모든 행에서 3~5번 열 추출 (3, 4, 5번) ---")
display(df.iloc[:, 3:6].head())

# iloc 슬라이싱
print("\n--- 모든 행에서 짝수 번호 열만 추출 (0, 2, 4번) ---")
display(df.iloc[:, 0:6:2].head())

# loc 슬라이싱
print("\n--- 모든 행에서 'year', 'pop' 열 추출 ---")
display(df.loc[:, ['year', 'pop']].head())

# iloc와 range() 함수 활용
print("\n--- 처음 5개 열 (0, 1, 2, 3, 4) ---")
display(df.iloc[:, list(range(5))].head())
```

```python
# 6. 기초 통계 - 그룹화 (Groupby)

# 단일 열 그룹화
print("--- 연도별 기대수명 평균 ---")
display(df.groupby('year')['lifeExp'].mean().head())

# 다중 열 그룹화
print("\n--- 연도 + 대륙별 기대수명, GDP 평균 ---")
display(df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].mean().head())

# reset_index()
print("\n--- 계층 인덱스를 일반 열로 변환 ---")
display(df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].mean().reset_index().head())
```

```python
# 7. 기초 통계 - 고유 값 및 빈도 분석

# nunique()
print(f"--- 대륙별 국가 수(중복 제외) ---")
display(df.groupby('continent')['country'].nunique())

# value_counts()
print(f"\n--- 대륙별 데이터 행 수(빈도수) ---")
display(df['continent'].value_counts())
```
