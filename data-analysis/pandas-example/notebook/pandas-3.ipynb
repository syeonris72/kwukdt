```python
import pandas as pd
import os

# 'output' 디렉토리 연결
output_dir = '/content/drive/MyDrive/kwukdt/data-analysis/pandas-example/output'
os.makedirs(output_dir, exist_ok=True)
```

```python
scientists = pd.read_csv("/content/drive/MyDrive/kwukdt/data-analysis/data-collection/scientists.csv")
display(scientists)
```

```python
# 피클(Pickle)
# scientists 데이터프레임에서 Name 열 추출
scientists_names = scientists['Name']

# 시리즈(Series)를 피클로 저장
series_pickle_path = os.path.join(output_dir, 'scientists_names.pkl')
scientists_names.to_pickle(series_pickle_path)
print(f"시리즈를 '{series_pickle_path}'에 저장했습니다.")

# 데이터프레임(DataFrame)을 피클로 저장
df_pickle_path = os.path.join(output_dir, 'scientists_df.pkl')
scientists.to_pickle(df_pickle_path)
print(f"데이터프레임을 '{df_pickle_path}'에 저장했습니다.")
```

```python
# 시리즈 피클 불러오기
loaded_names = pd.read_pickle(series_pickle_path)
print("불러온 시리즈 (처음 5개):")
display(loaded_names.head())

# 데이터프레임 피클 불러오기
loaded_scientists = pd.read_pickle(df_pickle_path)
print("\n불러온 데이터프레임 (처음 5개):")
display(loaded_scientists.head())
```

```python
# CSV/TSV 파일
# 데이터프레임을 CSV로 저장 (인덱스 제외)
csv_path = os.path.join(output_dir, 'scientists.csv')
scientists.to_csv(csv_path, index=False)
print(f"데이터프레임을 '{csv_path}'에 CSV 형식으로 저장했습니다.")

# 데이터프레임을 TSV로 저장 (탭 구분자, 인덱스 제외)
tsv_path = os.path.join(output_dir, 'scientists.tsv')
scientists.to_csv(tsv_path, sep='\t', index=False)
print(f"데이터프레임을 '{tsv_path}'에 TSV 형식으로 저장했습니다.")
```

```python
# CSV 파일 불러오기
loaded_scientists_csv = pd.read_csv(csv_path)
print("불러온 CSV 데이터프레임 (처음 5개):")
display(loaded_scientists_csv.head())

# TSV 파일 불러오기 (탭 구분자 지정)
loaded_scientists_tsv = pd.read_csv(tsv_path, sep='\t')
print("\n불러온 TSV 데이터프레임 (처음 5개):")
display(loaded_scientists_tsv.head())
```

```python
# 딕셔너리(Dictionary)
from pprint import pprint

# DataFrame을 'index' orient로 딕셔너리로 변환
# 현재 scientists 데이터프레임의 'Born'과 'Died'는 문자열
scientists_dict = scientists.to_dict(orient='index')
print("DataFrame을 딕셔너리로 변환 (처음 2개 항목):")
for i, (idx, row_dict) in enumerate(scientists_dict.items()):
    if i >= 2:
        break
    pprint(f"인덱스 {idx}: {row_dict}")
```

```python
# 딕셔너리를 다시 DataFrame으로 변환
loaded_scientists_dict = pd.DataFrame.from_dict(scientists_dict, orient='index')

# 'Born'과 'Died' 열이 문자열로 로드되었을 수 있으므로 datetime 객체로 다시 변환
loaded_scientists_dict['Born'] = pd.to_datetime(loaded_scientists_dict['Born'])
loaded_scientists_dict['Died'] = pd.to_datetime(loaded_scientists_dict['Died'])

print("딕셔너리에서 불러온 데이터프레임 (처음 5개):")
display(loaded_scientists_dict.head())
```

```python
# JSON(JavaScript Object Notation)
import io

# 데이터프레임을 JSON 문자열로 변환 (각 행을 딕셔너리로 표현, 2칸 들여쓰기)
# scientists 데이터프레임에는 'DiscoverDate' 열이 없으므로, 해당 열에 대한 특별한 처리 생략
scientists_json_string = scientists.to_json(orient='records', indent=2)
print("DataFrame을 JSON 문자열로 변환 (처음 일부):")
print(scientists_json_string[:500] + "...") # 문자열의 일부만 출력하여 가독성 높임
```

```python
# JSON 문자열을 데이터프레임으로 변환
loaded_scientists_json = pd.read_json(
    io.StringIO(scientists_json_string),
    orient='records'
)

# 'Born'과 'Died' 열이 문자열로 로드되었을 수 있으므로 datetime 객체로 다시 변환
loaded_scientists_json['Born'] = pd.to_datetime(loaded_scientists_json['Born'])
loaded_scientists_json['Died'] = pd.to_datetime(loaded_scientists_json['Died'])

print("\nJSON에서 불러온 데이터프레임 (처음 5개):")
display(loaded_scientists_json.head())
```
