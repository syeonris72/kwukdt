```python
gapminder = pd.read_csv("/content/drive/MyDrive/kwukdt/data-analysis/data-collection/gapminder.tsv", sep='\t')
display(gapminder)
```

```python
# rename을 사용하여 칼럼(Column) 변경
gapminder = gapminder.rename(columns={
    'country': '국가',
    'continent': '대륙',
    'year': '연도',
    'lifeExp': '기대수명',
    'pop': '인구',
    'gdpPercap': '1인당 GDP'
})
display(gapminder.head())
```

```python
print(f"DataFrame 구성: {gapminder.shape}")
print(f"\nDataFrame 열: {gapminder.columns}")
print(f"\n===== DataFrame 데이터 유형 =====\n{gapminder.dtypes}")
print(f"\n===== DataFrame 정보 =====")
gapminder.info()
print(f"\n===== 처음 5개 행 =====")
display(gapminder.head())
print(f"\n===== 마지막 5개 행 =====")
display(gapminder.tail())
```

```python
# groupby를 사용하여 '국가' 별 '기대수명' 평균 (소수점 둘째자리 반올림)
display(gapminder.groupby('국가')['기대수명'].mean().round(2))
```

```python
# '대륙'별 '국가'의 빈도수 (value_counts() 사용)
display(gapminder.groupby('대륙')['국가'].value_counts())
```
